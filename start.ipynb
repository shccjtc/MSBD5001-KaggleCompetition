```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
#Performing training
train = pd.read_csv('./msbd5001-fall2019/train.csv')
train = train.sample(frac =1,random_state=10)
```

```python
train['is_free'] = train['is_free']-0
train['purchase_days'] = (pd.to_datetime('2019-9-1 00:00:00')-pd.to_datetime(train['purchase_date'])).dt.days
train['release_days'] = (pd.to_datetime('2019-9-1 00:00:00')-pd.to_datetime(train['release_date'])).dt.days
# train = train.drop(['purchase_date','release_date','id','genres','categories','tags'],axis=1)
train = train.drop(['purchase_date','release_date','id'],axis=1)
print(train.shape)
```

```python
train['genres'] = train['genres'].str.split(',')
train['categories'] = train['categories'].str.split(',')
train['tags'] = train['tags'].str.split(',')
# train = train.drop(['genres','categories','tags'],axis=1)
print(train.shape)
```

```python
train
```

```python
gendict = {}
catdict = {}
tagdict = {}
for index,row in train.iterrows():
    genres = row['genres']
    cat = row['categories']
    tag = row['tags']
    for i in genres:
        gendict[i] = 1
    for i in cat:
        catdict[i] = 1
    for i in tag:
        tagdict[i] = 1

        
```

```python
for i in gendict.keys():
    train['g'+i] = 0

for i in catdict.keys():
    train['c'+i] = 0

for i in tagdict.keys():
    train['t'+i] = 0
print(train.shape)
```

```python
# train.loc[0,'price']=1
for i in range(train.shape[0]):
    l = train.loc[i,'genres']
    for j in l:
        train.loc[i,'g'+j] = 1
        
for i in range(train.shape[0]):
    l = train.loc[i,'categories']
    for j in l:
        train.loc[i,'c'+j] = 1
        
for i in range(train.shape[0]):
    l = train.loc[i,'tags']
    for j in l:
        train.loc[i,'t'+j] = 1

print(train.shape)
```

```python
train = train.dropna().drop(['genres','categories','tags'],axis=1).reset_index(drop=True)
print(train.shape)
```

```python
label = train['playtime_forever']
train = train.drop(['playtime_forever'],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train,label,random_state=10, test_size=0.25)
```

```python
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
model_cb = CatBoostRegressor()
```

```python
model_lgb = LGBMRegressor()

#Defining the list parameters
param_dist = {"learning_rate" : [0.0001,0.0005,0.001,0.01],
              "num_leaves": [7,10,15,31],
              "min_child_weight": [0.0001,0.001,0.005,0.01],
              "min_child_samples":[10,20,30,50] ,
              "num_iterations": [100,200,300,500]
             }

#Performing finetuing
grid_search = GridSearchCV(model_lgb, n_jobs=-1, param_grid=param_dist, cv = 3, verbose = 3)
grid_search.fit(X_train,y_train)
grid_search.best_estimator_

# 0.01, 0.05, 0.1 ->0.01  3,7,10,15,31,50 ->3  0.01,0.1,1->0.01  10,20,30->30
#0.001  7  0.001  20  300
#0.0005 7 0.0001 20 500
```

```python

model_cb.fit(X_train,y_train)
```

```python
model_lgb = LGBMRegressor(learning_rate = 0.005, num_leaves = 10,min_child_weight=0.001 ,min_child_samples = 20,num_iterations=1000)
model_lgb.fit(X_train,y_train)
```

```python
from sklearn import metrics
def auc(m, train, test): 
    return (metrics.mean_squared_error(y_train,m.predict(train)),
                            metrics.mean_squared_error(y_test,m.predict(test)))

print(auc(model_cb,X_train,X_test))
print(auc(model_lgb,X_train,X_test))
```
